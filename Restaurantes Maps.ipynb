```python
num=len(browser.find_elements("xpath","//div[@class='Nv2PK THOPZb CpccDe']"))
print(num)

list_pais=[]
for i in range(0,num):
    titular=browser.find_elements("xpath","//div[@class='Nv2PK THOPZb CpccDe']")[i].text
    list_pais.append(titular)

dfMaps=pd.DataFrame(list_pais)
dfMaps['origen']='Maps'
dfMaps
```

```python
pd.set_option('display.max_rows', 4)

fecha=datetime.now().strftime("%d-%m-%Y")
hora=datetime.now().strftime("%H")
dfMaps_Columns = dfMaps[0].str.split('\n', expand=True)
dfMaps_Columns[['rating','reviews']] = dfMaps_Columns[1].str.split('(',1,expand=True)
dfMaps_Columns[['reviews','price']]  = dfMaps_Columns['reviews'].str.split("\)",1,expand=True)
dfMaps_Columns[['price']]            = dfMaps_Columns[['price']].replace('·',value='',regex=True)
dfMaps_Columns['ciudad']           = ciudad
dfMaps_Columns['comida']           = comida
dfMaps_Columns['zona']             = barrio
dfMaps_Columns['comunidad']        = comunidad
dfMaps_Columns[0] = dfMaps_Columns[0].replace(u"\U0001F354",value="", regex=True)
dfMaps_Columns['nombre'] = dfMaps_Columns[0].replace(u"\U0001F35F",value="", regex=True)
dfMaps_Columns['fecha']=fecha
dfMaps_Columns['hora']=hora
dfMaps_Columns['clave']=dfMaps_Columns['nombre']+' '+dfMaps_Columns[2]+' '+dfMaps_Columns['fecha']

dfMaps_Columns
```

```python
col_list = ['nombre', 1,2,'rating','reviews','price',"comunidad","ciudad","comida","zona","fecha","hora","clave"] #3,4,5, excluimos esas dos columnas
dfMaps_Columns_selected = dfMaps_Columns[col_list]
dfMaps_Columns_selected
```

```python
import pymysql
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://{user}:{pw}@127.0.0.1/{db}"
                       .format(user="root",
                               pw="MariaDB1.2.3",
                               db="restaurants_maps"))

dfMaps_Columns_selected.to_sql('restaurants', con=engine,if_exists='append', chunksize=1000)    
```
