```python
from geopy.geocoders import Nominatim
import requests
import pandas as pd

# Tu clave de API de Google Maps
api_key = 'AIzaSyAhlwCCNhis1CT_JiWMIThGN7DG2lhMMFY'

# URL de la API de Places
url_search = 'https://maps.googleapis.com/maps/api/place/textsearch/json'
url_details = 'https://maps.googleapis.com/maps/api/place/details/json'

# Crear una lista vacía para almacenar los resultados
results = []

# Crear una lista con los barrios de Madrid
barrios = ['Chamartin', 'Tetuán', 'Chamberí', 'Fuencarral-El Pardo', 'Moncloa-Aravaca', 'Latina', 'Carabanchel', 'Usera', 'Puente de Vallecas', 'Moratalaz', 'Ciudad Lineal', 'Hortaleza', 'Villaverde', 'Villa de Vallecas', 'Vicálvaro', 'San Blas-Canillejas', 'Barajas']

# Crear una instancia de geolocalizador
geolocator = Nominatim(user_agent="geoapiExercises")

# Recorrer cada barrio
for barrio in barrios:
    # Geolocalizar el barrio
    location = geolocator.geocode(barrio + ", Madrid, España")
    # Obtener las calles del barrio
    calles = geolocator.reverse(location.point, exactly_one=False, timeout=10)
    # Recorrer cada calle
    for result in calles:
        if 'road' in result.raw['address']:
            calle = result.raw['address']['road']
        else:
            calle = ''
        results_calle = []
        # Realizar la búsqueda
        params_search = {
            'query': 'clínicas de fisioterapia en ' + calle,
            'key': api_key
        }
        response_search = requests.get(url_search, params=params_search)
        data = response_search.json()
        # Recorrer los resultados y obtener información
        for result in data['results']:
            place_id = result['place_id']
            params_details = {
                'place_id': place_id,
                'key': api_key
            }
            response_details = requests.get(url_details, params=params_details)
            data_details = response_details.json()
            if 'formatted_phone_number' in data_details['result']:
                phone = data_details['result']['formatted_phone_number']
            else:
                phone = None
            results_calle.append({
                'name': result['name'],
                'address': result['formatted_address'],
                'phone': phone,
                'barrio': barrio
            })
        results.extend(results_calle)


# Crear un dataframe con los resultados
df = pd.DataFrame(results)

# Guardar el dataframe en un archivo excel
#df.to_excel('C:/Users/vir_d/Desktop/clinicas_fisioterapia_madrid.xlsx',index=False)

```
